```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

Initiate empty lists.

```python
electiontype = []
state = []
date = []
party = []
candidate = []
votes = []
votes_pct = []
delegates = []
```

Prepare for web scraping.

```python
header = {'User-Agent':\
          'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/60.0.3112.113 Safari/537.36'}

url = 'http://www.politico.com/mapdata-2016/2016-election/primary/results/map/president/'

soup = requests.get(url, headers=header)
soup = BeautifulSoup(soup.text, "html5lib")
```

Perform web scraping and store the data into lists.

```python
for item in soup.findAll('article', class_='timeline-group'):
    dem = item.findAll('tr', class_='type-democrat')
    gop = item.findAll('tr', class_='type-republican')
    
    # current electiontype
    if(item.h4.text.split()[1][0] == 'P'): 
        currElecType = 'Primary'
    else:
        currElecType = 'Caucus'
        
    # current state
    currState = item.get('id')[-2:]
    
    # current date
    currDate = item.li['data-dt']
    
    for d in dem:
        electiontype.append(currElecType)
        state.append(currState)
        date.append(currDate)
        party.append('DEM')
        candidate.append(d.find('th', class_='results-name').text)
        votes.append(d.find('td', class_='results-popular').text)
        votes_pct.append(d.find('td', class_='results-percentage').text)
        try:
            delegates.append(d.find('td', class_='delegates-cell').text)
        except:
            delegates.append('')
        
    for g in gop:
        electiontype.append(currElecType)
        state.append(currState)
        date.append(currDate)
        party.append('GOP')
        candidate.append(g.find('th', class_='results-name').text)
        votes.append(g.find('td', class_='results-popular').text)
        votes_pct.append(g.find('td', class_='results-percentage').text)
        try:
            delegates.append(g.find('td', class_='delegates-cell').text)
        except:
            delegates.append('')  
               
```

Aggregate results into pandas dataframe.

```python
df_results = pd.DataFrame({'electiontype': electiontype,\
                          'state': state,\
                          'date': date,\
                          'party': party,\
                          'candidate': candidate,\
                          'votes': votes,\
                          '%votes': votes_pct,\
                          'delegates': delegates})
```

```python
df_results = df_results.set_index(['electiontype','state','date','party'])
```
